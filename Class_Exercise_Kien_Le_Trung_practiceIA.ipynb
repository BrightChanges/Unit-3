# **Development**

Following on the UML class diagram on **Fig. 1**, the class definition is shown below. Note: the class attribute order_no  is a varaible shared among all the objects created from this class. Keeping track the order_no is like  tracking id of each breakfast order, preventing the system to mistakenly processing the wrong breakfast order.

```
class Breakfast_orders():

  # This is a class's attribute. Common to all the objects
  order_no: int = 0

  # Initializer of the class. This is also called the constructor method
  def __init__(self, food_list, price, delivery_address):
    # These are the attribues for each object. the key word self refers to particular object.
    self.food_list :str = food_list
    self.price :int = price #price in 
    self.delivery_address :str = delivery_address

#when an object of this class is instantiated:
    Breakfast_orders.order_no += 1

# this is a method to get general information about the Breakfast_orders:
  def get_order_info(self)-> str:
     #This method returns the basic information for the motocycle as a str
    return f"<Information about this breakfast order: food :{self.food_list}, price:{self.price}, delivery_address:{self.delivery_address}>"

  def remove_breakfast_order(self):
    # this method removes a breakfast order
    Breakfast_orders.order_no -= 1
    # deleting the object, which here is referenced by the word self
    del self
```

```
#Create a test object of the class Breakfast Orders:
order_from_tom = Breakfast_orders(food_list="3 eggs, 5 pancakes, 4 sausages", price=300, delivery_address="ISAK'Big Cafeteria")
print(order_from_tom.get_order_info())
#Check tracking id of this order:
print(f"Tracking id of this breakfast order: {Breakfast_orders.order_no}")

```

```
order_from_tom.remove_breakfast_order()
print(f"Tracking id of this breakfast order: {Breakfast_orders.order_no}, if tracking id is 0, the system hasn't received any breakfast order yet.")
```
